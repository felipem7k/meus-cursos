```python
import numpy as np

class Pilha:
    def __init__(self, capacidade):
        self.capacidade = capacidade
        self.ultima_posicao = -1
        self.valores = np.empty(self.capacidade, dtype=int)

    def vazia(self):
        return self.ultima_posicao == -1

    def cheia(self):
        return self.ultima_posicao + 1 == self.capacidade

    def topo(self):
        if self.vazia():
            print("Pilha vazia!")
            return

        print(self.valores[self.ultima_posicao])

    def empilhar(self, valor):
        if self.cheia():
            print(f"Valor {valor} não inserido. Pilha já está cheia!")
            return

        prox_posicao = self.ultima_posicao + 1

        self.ultima_posicao = prox_posicao
        self.valores[prox_posicao] = valor

    def desempilhar(self):
        if self.ultima_posicao <= -1:
            return
        
        self.valores = self.valores[:-1]
        self.ultima_posicao -= 1
```

```python
pilha = Pilha(5)

pilha.empilhar(10)
pilha.empilhar(20)
pilha.empilhar(49)
pilha.topo()
pilha.desempilhar()
pilha.topo()

if pilha.vazia():
    print("A pilha está vazia, adicione algum valor para começar...")
elif pilha.cheia():
    print("A pilha está cheia!")
else:
    print(f"Existem {pilha.ultima_posicao + 1} valores na pilha, você pode adicionar mais valores...")
```
