```python
def soma(total: int) -> int:
    resultado: int = 1
    atual: int = 2
    while (atual <= total):
        resultado += atual
        atual += 1

    return resultado

print(soma(10))
```

```python
import timeit

def soma(n):
    soma = 0
    for i in range(n + 1):
        soma += 1

    return soma

def soma_otimizada(n):
    return (n * (n + 1)) / 2

tempo_inicial = timeit.default_timer()
soma_otimizada(10000000000)
tempo_final = timeit.default_timer()

print(tempo_final - tempo_inicial)
```

```python
import timeit

quantidade: int = 10000000
lista: list = []

tempo_inicial = timeit.default_timer()

for i in range(1, quantidade + 1):
    lista.append(i)

tempo_final = timeit.default_timer()

print(tempo_final - tempo_inicial)
```

```python
import numpy as np
import matplotlib.pyplot as plt
n = np.linspace(1, 10, 100)
labels = ["Constante", "Logaritmico", "Linear"]
big_o = [np.ones(n.shape), np.log(n), n]

plt.figure(figsize=(5, 4))
plt.ylim(0, 10)

for i in range(len(big_o)):
    plt.plot(n, big_o[i], label=labels[i])

plt.legend()
plt.ylabel("Tempo")
plt.xlabel("n")
plt.show()
```
