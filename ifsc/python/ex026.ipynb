```python
class VetorOrdenado:
    def __init__(self, capacidade):
        self.capacidade = capacidade
        self.ultima_posicao = -1
        self.valores = np.empty(self.capacidade, dtype=int)

    def imprime(self):
        if self.ultima_posicao == -1:
            print("Vetor vazio!")
            return
        
        for i in range(self.ultima_posicao + 1):
            print(f"{i} - {self.valores[i]}")

    def incrementa(self, index, valor):
        if index > self.capacidade - 1:
            return

        atual = self.valores[index]

        self.valores[index] = valor

        self.incrementa(index + 1, atual)

    def maior_mais_proximo(self, alvo):
        for index in range(self.ultima_posicao + 1):
            valor_atual = self.valores[index]
            if valor_atual > alvo:
                return index

        return -1

    def coletar_metade(self, iniciar_em, finalizar_em):
        if finalizar_em > 0:
            if iniciar_em >= finalizar_em:
                return None
        
            return math.floor((iniciar_em + finalizar_em) / 2)
        
        return finalizar_em

    def procura_metade(self, iniciar_em, finalizar_em, alvo):
        index_atual = self.coletar_metade(iniciar_em, finalizar_em)

        if not index_atual:
            return -1

        valor_metade = self.valores[index_atual]
        if valor_metade == alvo:
            return index_atual
        elif valor_metade < alvo:
            if index_atual >= self.ultima_posicao - 1:
                index_penultimo = self.procura_metade(index_atual, finalizar_em, alvo)
                if self.valores[index_penultimo] == alvo:
                    return index_penultimo
                elif self.valores[index_penultimo + 1] == alvo:
                    return index_penultimo + 1
                return -1
            
            return self.procura_metade(index_atual, finalizar_em, alvo)
        elif valor_metade > alvo:
            if index_atual >= self.ultima_posicao:
                return self.ultima_posicao
            
            return self.procura_metade(iniciar_em, index_atual, alvo)

    def pesquisa(self, alvo):
        return self.procura_metade(0, self.ultima_posicao, alvo)

    def inclui(self, valor):
        if self.ultima_posicao >= self.capacidade - 1:
            print(f"Valor {valor} não foi inserido. Capacidade máxima atingida!")
            return

        self.ultima_posicao += 1

        index = self.maior_mais_proximo(valor)
        if index == -1:
            index = self.ultima_posicao

        self.incrementa(index, valor)

    def percorre(self, index):
        if index >= self.capacidade - 1:
            return

        proximo_index = index + 1
        self.valores[index] = self.valores[proximo_index]
        self.percorre(proximo_index)

    def exclui(self, valor):
        if self.ultima_posicao == -1:
            return

        index_lista = self.pesquisa(valor)

        if index_lista <= -1:
            return

        self.percorre(index_lista)

        self.ultima_posicao -= 1



```

```python
import timeit
vetor = VetorNaoOrdenado(5)

tempo_inicial = timeit.default_timer()

vetor.inclui(4) 
vetor.inclui(6)
vetor.inclui(8)
vetor.inclui(12)
vetor.inclui(5)
vetor.inclui(9)

vetor.imprime()
print(f"locais com numero 12 {vetor.pesquisa(12)}")

vetor.exclui(6)

vetor.imprime()

tempo_final = timeit.default_timer()

print(f"Tempo total de desempenho: {tempo_final - tempo_inicial}")
```
