```python
class VetorOrdenado:
    def __init__(self, capacidade):
        self.capacidade = capacidade
        self.ultima_posicao = -1
        self.valores = np.empty(self.capacidade, dtype=int)

    def imprime(self):
        if self.ultima_posicao == -1:
            print("Vetor vazio!")
            return
        
        for i in range(self.ultima_posicao + 1):
            print(f"{i} - {self.valores[i]}")

    def incrementa(self, index, valor):
        if index > self.capacidade - 1:
            return

        atual = self.valores[index]

        self.valores[index] = valor

        self.incrementa(index + 1, atual)

    def maior_mais_proximo(self, alvo):
        for index in range(self.ultima_posicao + 1):
            valor_atual = self.valores[index]
            if valor_atual > alvo:
                return index

        return -1

    def pesquisa(self, alvo):
        for index in range(self.ultima_posicao + 1):
            valor_atual = self.valores[index]
            if valor_atual > alvo:
                break

            if valor_atual == alvo:
                return index
        return -1

    def inclui(self, valor):
        if self.ultima_posicao >= self.capacidade - 1:
            print(f"Valor {valor} não foi inserido. Capacidade máxima atingida!")
            return

        self.ultima_posicao += 1

        index = self.maior_mais_proximo(valor)
        if index == -1:
            index = self.ultima_posicao

        self.incrementa(index, valor)

    def percorre(self, index):
        if index >= self.capacidade - 1:
            return

        proximo_index = index + 1
        self.valores[index] = self.valores[proximo_index]
        self.percorre(proximo_index)

    def exclui(self, valor):
        if self.ultima_posicao == -1:
            return

        index_lista = self.pesquisa(valor)

        if index_lista <= -1:
            return

        self.percorre(index_lista)

        self.ultima_posicao -= 1



```
