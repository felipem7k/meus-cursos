```python
import numpy as np

class Fila:
    def __init__(self, capacidade):
        self.capacidade = capacidade
        self.primeira_posicao = 0
        self.ultima_posicao = capacidade
        self.posicao_atual = -1
        self.qtd_itens = 0
        self.valores = np.empty(self.capacidade, dtype=int)

    def vazia(self):
        return self.qtd_itens == 0

    def cheia(self):
        return self.qtd_itens == self.capacidade

    def atual(self):
        if self.vazia():
            return None

        return self.valores[self.primeira_posicao]

    def mostrar_atual(self):
        print(self.atual() or "Fila vazia!")

    def empilhar(self, valor):
        if self.cheia():
            print(f"Valor {valor} não inserido. Fila já está cheia!")
            return

        self.posicao_atual += 1
        if self.posicao_atual > self.capacidade - 1:
            self.posicao_atual = 0


        self.valores[self.posicao_atual] = valor
        self.qtd_itens += 1

    def desempilhar(self):
        if self.vazia():
            return
        
        ultimo_index = self.capacidade - 1

        self.primeira_posicao += 1
        if self.primeira_posicao > ultimo_index:
            self.primeira_posicao = 0

        self.ultima_posicao += 1
        if self.ultima_posicao > ultimo_index:
            self.ultima_posicao = 0

        self.qtd_itens -= 1

        if self.vazia():
            self.posicao_atual = self.primeira_posicao
        
```

```python
fila = Fila(3)

fila.mostrar_atual()

fila.empilhar(2)
fila.empilhar(5)

fila.mostrar_atual()

fila.desempilhar()

fila.mostrar_atual()

fila.empilhar(10)
fila.empilhar(30)
fila.desempilhar()
fila.desempilhar()
fila.empilhar(40)
fila.empilhar(25)
fila.empilhar(50)

fila.mostrar_atual()

fila.desempilhar()

fila.mostrar_atual()
```
