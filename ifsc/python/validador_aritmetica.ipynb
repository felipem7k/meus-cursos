```python
import numpy as np

class Pilha:
    def __init__(self, capacidade):
        self.capacidade = capacidade
        self.ultima_posicao = -1
        self.valores = np.empty(self.capacidade, dtype=str)

    def vazia(self):
        return self.ultima_posicao == -1

    def cheia(self):
        return self.ultima_posicao + 1 == self.capacidade

    def topo(self):
        if self.vazia():
            return -1

        return self.valores[self.ultima_posicao]

    def empilhar(self, valor):
        if self.cheia():
            print(f"Valor {valor} não inserido. Pilha já está cheia!")
            return

        prox_posicao = self.ultima_posicao + 1

        self.ultima_posicao = prox_posicao
        self.valores[prox_posicao] = valor

    def desempilhar(self):
        if self.ultima_posicao <= -1:
            return
        
        self.valores = self.valores[:-1]
        self.ultima_posicao -= 1
```

```python
class ExpressaoAritmetica:
    def __init__(self, expressao):
        self.expressao = expressao

        self.__caracteres = Pilha(len(self.expressao))

    def __casa(self, digito):
        match (digito):
            case ")":
                return "("
            case "]":
                return "["
            case "}":
                return "{"
            case _:
                return None

    def __abre(self, digito):
        self.__caracteres.empilhar(digito)

    def __fecha(self, digito):
        casal = self.__casa(digito)

        if not casal:
            raise Exception(f"Erro no digito '{digito}'!")
        
        if self.__caracteres.topo() != casal:
            print(self.__caracteres.topo(), casal)
            raise Exception(f"Erro, fechamento '{digito}' sem abertura!")
        
        self.__caracteres.desempilhar()

    def __valida(self, digito):
        if digito in ["(", "[", "{"]:
            self.__abre(digito)
        elif digito in [")", "]", "}"]:
            self.__fecha(digito)

    def verifica(self):
        for digito in self.expressao.strip():
            self.__valida(digito)

        if not self.__caracteres.vazia():
            print(f"Erro, faltando fechamento de '{self.__caracteres.topo()}'")
```

```python
expressao = ExpressaoAritmetica("2 + (3 + 3 - (2 - 1))")

expressao.verifica()
```
